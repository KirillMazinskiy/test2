```python
df = df_bets.merge(df_events,  on='event_id', how='inner')
df
```

```python
df_clients = (df[
    (df['sport'] == 'E-Sports')
    &
    (df['amount']  >= 10.0) # bonus_action
    &
    (df['event_stage']  == 'Prematch')
    &
    (((df['bet_type']  == 'Ordinar') & (df['accepted_bet_odd']  >= 1.5)) | (df['bet_type'] == 'Express')) # Not System
    &
    (~df['result'].isin(['Cashout','TechnicalReturn','Return'])) # Not Cashout','TechnicalReturn','Return
    &
    (df['is_free_bet']  == False) # Not free_bet
    &
    (pd.to_datetime( df['create_time'])  >=  datetime(2022, 3, 14, 12))
    &
    (pd.to_datetime( df['settlement_time'])  <=  datetime(2022, 3, 15, 12))
  ])['player_id'].drop_duplicates()
```
